# **4Ô∏è‚É£ Data Cleaning & Transformation**

```python
# ‚úèÔ∏è Rename Columns for Readability
df = df.rename(columns={
    'transaction_id'   : 'Order_id',
    'transaction_qty'  : 'Quantity',
    'transaction_date' : 'Order_date',
    'transaction_time' : 'Time'
})
```

```python
# ‚ûï Add Total_Price Column
df['Total_Price'] = df['Quantity'] * df['unit_price']
```

```python
# üóëÔ∏è Drop Unnecessary Columns
df = df.drop(['store_id', 'product_id', 'product_detail'], axis=1)
```

```python
# üßπ Drop Missing Values
df = df.dropna()
```

```python
# üìÜ Convert Order_date to DateTime
df['Order_date'] = pd.to_datetime(df['Order_date'], errors='coerce')
```

## **5Ô∏è‚É£ Standardize Text Columns**

```python
# üé® Standardize string formatting
df['store_location']   = df['store_location'].str.strip().str.title()
df['product_category'] = df['product_category'].str.strip().str.title()
df['product_type']     = df['product_type'].str.strip().str.title()
```

# **6Ô∏è‚É£ Validate Cleaned Data**

```python
# ‚úÖ Quick sanity check
print("Final Shape:", df.shape)
print("Null Values:\n", df.isnull().sum())
df.head()
```

# **7Ô∏è‚É£  Save Cleaned Dataset**


```python
# üíæ Save cleaned dataset
df.to_csv('Cleaned_Coffee_Sales_Dataset.csv', index=False)

print("‚úÖ Data cleaning complete! Cleaned file saved as 'Cleaned_Coffee_Sales_Dataset.csv'")
```
